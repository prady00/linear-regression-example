## Scripts to predict the loaded passengers on the basis of history data using stastical linear regrassion model

```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
#Importing the dataset & hava a look on it
data = pd.read_csv("data/Forecast.csv")
data.head()
```

# Simple Linear Regression

As first step, we can create model having linear relationship between Booked & Loaded data values

```python
#Initializing the variables
X = data['Booked'].values.reshape(-1,1)
y = data['Loaded'].values.reshape(-1,1)
```

```python
#Ploting a graph to see the points
plt.figure(figsize=(16, 8))
plt.scatter(X, y, c='black')
plt.xlabel("Booked")
plt.ylabel("Loaded")
plt.show()
```

```python
#Splitting our dataset to Training and Testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Fitting Linear Regression to the training set
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
#predicting the Test set result
y_pred = reg.predict(X_test)
plt.figure(figsize=(16, 8))
plt.scatter(X, y, c='black')
plt.plot(
    X_test,
    y_pred,
    c='blue',
    linewidth=2
)
plt.xlabel("Booked")
plt.ylabel("Loaded")
plt.show()
```

```python
#Calculating the Coefficients
reg.coef_
```

```python
#Calculating the Intercept
reg.intercept_
```
